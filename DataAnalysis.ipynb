```python
# importing pandas library for data analysis
import pandas as pd
```

```python
# setting up for loading dataset
path = "/Users/shiva/Downloads/automobile/data.csv"
#loading the dataset using read_csv
df = pd.read_csv(path,header = None)
#printing the dataset 
df
```

```python
# showing Top 5 rows from the Dataset
df.head()
```

```python
# showing last 5 rows from the dataset
df.tail()
```

```python
# In the Above Dataset there are no columns name which is headers 
# Now we are replacing with headers 
# Columns Names in headers varriable in the form of list
headers = ["symboling","normalized-losses",
           "make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels","engine-location",
           "wheel-base","length",
           "width","height","curb-weight",
           "engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke",
           "compression-ratio","horsepower",
           "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#Length of the headers
len(headers)
```

```python
# Replacing headers Using df.columns
df.columns = headers
```
